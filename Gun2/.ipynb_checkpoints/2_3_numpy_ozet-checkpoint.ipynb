### Numpy Özet

```python
#örn:
import numpy as np
import random as rnd

#satırxsutunxpikselRengi (RGB)
# 2 satır 3 sutun kesişimlerin 3 eleman olsun.
uDizi = np.array([
                    [(0,0,0), (255,0,0), (255,255,0)],
                    [(0,0,255), (0,255,255), (255,255,255)]
                ])

uDizi.shape #(2, 3, 3)
uDizi.ndim #3 boyut
uDizi.dtype #ortak tip
#uDizi.astype(bool) #ortak tipi değiştirme
#------
print(uDizi.size) #18
uDizi = uDizi.reshape((1,6,3)) #boyutunu değiştirebiliriz

#Broadcast özelliği. bir işlem tüm elemanlara tek tek uygulanır.
uDizi = uDizi * rnd.randint(0,10)

#----
import matplotlib.pyplot as plt
plt.imshow(uDizi)
plt.show()
```

* Soru:
    1. Resimdeki diziyi oluşturalım.
    2. Truncu alanı getiren
    3. mavi alan
    4. Kırmızı
    5. Mavi Çizgili
    6. 11, 43, 51 sayılarını getirelim
    7. Hem 2'ye hem de 3'e tam bölünen sayıları getirelim.



```python
#1.
dizi = np.arange(0,6)

#np.vstack([dizi, dizi+10,dizi+20, dizi+30, dizi+40,dizi+50])
dizi = np.vstack([dizi+t for t in range(0,60,10)])

#-----
#2.
dizi[0, (3,4)]
dizi[0, 3:5]

#3.
dizi[4:, 4:]

#4.
dizi[:, 2]

#5.
dizi[2::2, 0::2]

#6.11, 43, 51
dizi[(1,4,5),(1,3,1)] #indexing yöntemi (r1,r2),(c1,c2)

#7.
filtre_3 = dizi%3==0
filtre_2 = dizi%2==0

filtre = filtre_3 & filtre_2 #np.logical_and ...or, not veya &,|,~

dizi[filtre]
```

```python
#Soru:
'''
    - 10x10'luk 1'lerden oluşan bir dizi tanımlayalım.
    - içinde 6x6'lık 8'lerden oluşan bir dizi olsun.
    - onun içi de 1'lerden oluşsun.
'''

img = np.ones((10,10), dtype=int) 
#img.astype(int)

img[2:-2, 2:-2] = 8
img[3:-3, 3:-3] = 1

img
```
