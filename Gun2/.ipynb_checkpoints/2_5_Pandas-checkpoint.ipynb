### Null (None, NaN, N/A) Deerler

* `isna` veya `isnull` hcreleri kontrol eder. None olanlar True gelir.
* `notna` ve `notnull` isna nn tersi
* `all()` hepsi True mu?, `any()` hi True var m?
* `fillna` ile null alanlara deer atarz.
* `dropna` ile nullar kaldrabiliriz.

Not: fillna ve dropna axis alr. Yaplan ilem orjinalini etkilemez.

```python
ekip.isna() #isnull()

#
#axis verilebilir
ekip.isna().any() #hi True var m? yani null var m?
ekip.isna().all() #tamam null m?

#tabloda hi null var m?
ekip.isna().any().any()
```

```python
ekip.Uretim.isna().sum() #retim kolonunda ka null var?

#hangileri null
ekip.Uretim.isna()
ekip.Uretim.isna().tolist()

#
ekip[ekip.Uretim.isna()] #tm kolonlar gelecekse
ekip.loc[ekip.Uretim.isna(), ["Ad","Cinsiyet","Uretim"]] #etiketler kullanlackasa
ekip.iloc[ekip.Uretim.isna().tolist(), 1:4] #tolist ileliste yapmak gerekiyor. sra numaras kullanlacaksa

#etiket zerinden slicer yaplabilir.
ekip.loc[ekip.Uretim.isna(), "Ad":"Uretim"]
```

```python
#Soru:
#fillna kullanadan NULL olan retimlere ortalama retimi basalm.
ekip.loc[ekip.Uretim.isna(), "Uretim"] = ekip.Uretim.mean()
#ekip[ekip.Uretim.isna()].Uretim
ekip
```

```python
#fillna
ekip = ekipEfor.copy()

#
#ekip.fillna("Bilinmiyor") #tm null hcrelere deer verir.

#
ekip.fillna({
                "Uretim":ekip.Uretim.mean(),
                #"Cinsiyet":"Bilinmiyor",
                "Cinsiyet":ekip.Cinsiyet.mode()[0],
                "DepId":-1
})
```

```python
#fillna
ekip = ekipEfor.copy()

#
#ekip.dropna(axis=1, how="any") #herhangi bir null varsa axisi kaldr
ekip.dropna(axis=1, how="all") #hepsi null ise kaldr

#
ekip.dropna(axis=0, how="all") #bo satr komple kaldr
```

#### Kolan Ekleme ve kaldrma

```python
hedef = 500
ekip["HedefDurumu"] = ekip.Uretim / hedef
ekip
```

```python
#eval expression yardmyla kolon retmemizi salayacak
buyumeOrani = 1.2

#@ ile dardaki deikeni arabiliriz.
ekip.eval("""
            GelecekYilUretim = Uretim * @buyumeOrani
            SonrakiYilUretim = GelecekYilUretim * @buyumeOrani
           """, inplace=True)

ekip
```

```python
#orjinalini etkilemez.
ekip.drop(["HedefDurumu","GelecekYilUretim"], axis=1) #kolonu kaldrr. 

#
ekip.drop(index="orum") #satr etiketi

#
ekip.drop(index="Bolu", columns="Ad") #satr ve kolon etiketi.
```

```python
#soru:
#query
'''
    1. Cinsiyet alan ve Uretim alan null olmayanlar
    2. Cinsiyet alan null olanlar
    3. Uretim alan null olanlar
    4. Uretim alan null olmayanlar
    5. retimi null olmayan erkek yeler
    6. Ortalama retimin zerinde olan kadn yeler
'''

ekip = ekipEfor.copy()
#1.
ekip.query("Cinsiyet == Cinsiyet & Uretim == Uretim") #None ile None eit mi sorusu False retir.

#2.
ekip.query("Cinsiyet != Cinsiyet")

#3.
ekip.query("Cinsiyet.isnull()") #isna()

#4.
ekip.query("Uretim.notna()") #notnull()

#5.
ekip.query("Uretim.notna() and Cinsiyet =='E'") 

#6.
ekip.query(f"Uretim > {ekip.Uretim.mean()}")
#--
ortalama_uretim = ekip.Uretim.mean()
ekip.query("Uretim > @ortalama_uretim")
```

#### filter
* etiketleri tarif etmemizi salyor.
* items, regex, like

```python
#soru:
#iinde im geen kolonlar gelsin.
#ekip.columns == "Ad"
ekip.columns.str.contains("im")

#
ekip.loc[:, ekip.columns.str.contains("im")]
ekip.iloc[:, ekip.columns.str.contains("im")]
```

```python
ekip.filter(items = ["Ad","Uretim"])

#
ekip.filter(like="e") #ieriside e geenler
ekip.filter(like = "o", axis=0) #satr etiketlerinde ara

#
ekip.filter(regex="im$") #im ile bitenler
ekip.filter(regex="^I") #I ile balayanlar
```

### Groupby ve Agg

```python
ekip.count() #axis bazl sonu verir.

#
ekip.groupby(by="Cinsiyet").count()

#zetleme fonkisyonlarnda oklu kullanacaksak agg ile belirtiriz.
ekip.groupby(by="Cinsiyet").agg(["sum","mean","count"])

#hangi kolonda ne yaplacak
df = ekip.groupby(by="Cinsiyet").agg({"Uretim":["min","max","sum","mean"], "Id":["min","max"]}) #hiyearik index
df 
```

```python
#hiyerarik indexlerdeki bir alana erimek
df = df.reset_index()

df
```
